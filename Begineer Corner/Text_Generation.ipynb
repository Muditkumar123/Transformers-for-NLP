```python
from transformers import pipeline,set_seed
import textwrap
import numpy as np
import matplotlib.pyplot as plt

from pprint import pprint
```

```python
!cat robert_frost.txt
```

```python
lines=[line.rstrip() for line in open('robert_frost.txt')] # stripping off the white space
lines=[line for line in lines if len(line)>0] #Stripping/Removing extra line
```

```python
gen=pipeline("text-generation")
```

```python
set_seed(42)
```

```python
lines[0]
```

```python
gen(lines[0])
```

```python
pprint(_)#print above in readable format
```

```python
pprint(gen(lines[0],max_length=20))
```

```python
pprint(gen(lines[0],max_length=20,num_return_sequences=3))#num_return_sequences=3 means 3 possible text generate krna 
```

```python
def wrap(x):
  return textwrap.fill(x,replace_whitespace=False,fix_sentence_endings=True)
```

```python
out=gen(lines[0],max_length=30)
print(wrap(out[0]['generated_text']))
```

```python
prev="Two roads diverged in a yellow wood, while the water flowed across the"+ \
"river, and were covered by thick slag, a deep green sea."
out=gen(prev + '\n' + lines[2],max_length=60)
print(wrap(out[0]['generated_text']))
```

```python
prev="Two roads diverged in a yellow wood, while the water flowed across"+\
"the river, and were covered by thick slag, a deep green sea."+\
"And be one"+\
"traveler, long I stood."

out=gen(prev + '\n' + lines[4],max_length=90)
print(wrap(out[0]['generated_text']))
```

```python
prompt="Neural networks with attention have been used with great success in natural language processing."

out=gen(prompt,max_length=300)
print(wrap(out[0]['generated_text']))
```
